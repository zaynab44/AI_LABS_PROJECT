```python
# Question 4: SVM on Digits Dataset (Handwritten Digit Recognition)

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train SVM with RBF kernel
svm_digits = SVC(kernel='rbf', gamma=0.001)
svm_digits.fit(X_train, y_train)

# Predict and evaluate
y_pred = svm_digits.predict(X_test)
print("Accuracy (SVM - Digits):", accuracy_score(y_test, y_pred))

# Visualize some misclassified samples
misclassified = np.where(y_pred != y_test)[0][:5]
plt.figure(figsize=(10, 4))
for i, idx in enumerate(misclassified):
    plt.subplot(1, 5, i + 1)
    plt.imshow(X_test[idx].reshape(8, 8), cmap='gray')
    plt.title(f"True: {y_test[idx]}, Pred: {y_pred[idx]}")
plt.tight_layout()
plt.show()
```
