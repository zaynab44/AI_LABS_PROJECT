```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Dataset
data = {
    'StudyHours': ['Low', 'High', 'High', 'Low', 'High'],
    'Attendance': ['Poor', 'Good', 'Poor', 'Good', 'Good'],
    'Result': ['Fail', 'Pass', 'Pass', 'Fail', 'Pass']
}
df = pd.DataFrame(data)

# Encode categorical data
encoder = LabelEncoder()
for col in df.columns:
    df[col] = encoder.fit_transform(df[col])

X = df[['StudyHours', 'Attendance']].values
y = df['Result'].values

# Deep Neural Network (3 hidden layers)
dnn_model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(6, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])
dnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
dnn_model.fit(X, y, epochs=150, verbose=0)

# Shallow ANN (1 hidden layer)
ann_model = Sequential([
    Dense(6, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])
ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X, y, epochs=150, verbose=0)

# Compare accuracy
dnn_acc = dnn_model.evaluate(X, y, verbose=0)[1]
ann_acc = ann_model.evaluate(X, y, verbose=0)[1]

print(f"DNN Accuracy: {dnn_acc:.2f}")
print(f"Shallow ANN Accuracy: {ann_acc:.2f}")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Dataset
data = {
    'StudyHours': ['Low', 'High', 'High', 'Low', 'High'],
    'Attendance': ['Poor', 'Good', 'Poor', 'Good', 'Good'],
    'Result': ['Fail', 'Pass', 'Pass', 'Fail', 'Pass']
}
df = pd.DataFrame(data)

# Encode categorical data
encoder = LabelEncoder()
for col in df.columns:
    df[col] = encoder.fit_transform(df[col])

X = df[['StudyHours', 'Attendance']].values
y = df['Result'].values

# Deep Neural Network (3 hidden layers)
dnn_model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(6, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])
dnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
dnn_model.fit(X, y, epochs=150, verbose=0)

# Shallow ANN (1 hidden layer)
ann_model = Sequential([
    Dense(6, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])
ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X, y, epochs=150, verbose=0)

# Compare accuracy
dnn_acc = dnn_model.evaluate(X, y, verbose=0)[1]
ann_acc = ann_model.evaluate(X, y, verbose=0)[1]

print(f"DNN Accuracy: {dnn_acc:.2f}")
print(f"Shallow ANN Accuracy: {ann_acc:.2f}")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Load dataset
iris = load_iris()
X = iris.data
y = to_categorical(iris.target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model A: ReLU activation
relu_model = Sequential([
    Dense(16, activation='relu', input_shape=(4,)),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])
relu_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
relu_model.fit(X_train, y_train, epochs=150, verbose=0)
relu_acc = relu_model.evaluate(X_test, y_test, verbose=0)[1]

# Model B: tanh activation
tanh_model = Sequential([
    Dense(16, activation='tanh', input_shape=(4,)),
    Dense(12, activation='tanh'),
    Dense(8, activation='tanh'),
    Dense(3, activation='softmax')
])
tanh_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
tanh_model.fit(X_train, y_train, epochs=150, verbose=0)
tanh_acc = tanh_model.evaluate(X_test, y_test, verbose=0)[1]

print(f"ReLU Model Accuracy: {relu_acc:.2f}")
print(f"Tanh Model Accuracy: {tanh_acc:.2f}")
```

```python
from tensorflow.keras.datasets import mnist

# Load MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28*28) / 255.0
X_test = X_test.reshape(-1, 28*28) / 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Example: Different batch sizes
def build_and_train(batch_size, neurons):
    model = Sequential([
        Dense(neurons, activation='relu', input_shape=(784,)),
        Dense(neurons//2, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=5, batch_size=batch_size, verbose=0)
    acc = model.evaluate(X_test, y_test, verbose=0)[1]
    return acc

for batch in [32, 64, 128]:
    acc = build_and_train(batch, 256)
    print(f"Batch Size {batch}  Accuracy: {acc:.2f}")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout

# Overfitting model (no regularization)
overfit_model = Sequential([
    Dense(512, activation='relu', input_shape=(784,)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
overfit_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
overfit_model.fit(X_train, y_train, epochs=20, batch_size=128, verbose=0)
overfit_acc = overfit_model.evaluate(X_test, y_test, verbose=0)[1]

# Regularized model (Dropout + EarlyStopping)
reg_model = Sequential([
    Dense(512, activation='relu', input_shape=(784,)),
    Dropout(0.3),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
reg_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reg_model.fit(X_train, y_train, epochs=20, batch_size=128, validation_split=0.2, callbacks=[early_stop], verbose=0)
reg_acc = reg_model.evaluate(X_test, y_test, verbose=0)[1]

print(f"Without Regularization Accuracy: {overfit_acc:.2f}")
print(f"With Regularization Accuracy: {reg_acc:.2f}")
```
