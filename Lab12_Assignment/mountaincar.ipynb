```python
import os
os.environ["SDL_VIDEODRIVER"] = "windows"   # change to "x11" if on Linux

import gymnasium as gym
import pygame

env = gym.make("MountainCar-v0", render_mode="human")

font = None
scores = []

for episode in range(1, 51):
    state, info = env.reset()
    done = False
    score = 0
    steps = 0

    while not done:

        print("State:", state)   # shows position and velocity

        # Rule-based policy
        if state[1] > 0:
            action = 2
        else:
            action = 0

        state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        score += reward
        steps += 1

        if font is None:
            pygame.font.init()
            font = pygame.font.SysFont("Arial", 24)

        surface = pygame.display.get_surface()
        text = font.render(f"Episode: {episode} | Score: {int(score)}", True, (0, 0, 255))
        surface.blit(text, (200, 20))

        pygame.display.update()
        pygame.time.delay(20)

    scores.append(score)
    print(f"Episode {episode} Score: {score}, Steps: {steps}")

print("Best Score:", max(scores))

env.close()
pygame.quit()
```
