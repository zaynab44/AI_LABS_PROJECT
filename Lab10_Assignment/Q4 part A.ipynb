```python
import numpy as np
import matplotlib.pyplot as plt

# Points and initial centroids
points = {
    "P1": (1,3), "P2": (2,2), "P3": (5,8), "P4": (8,5), "P5": (3,9),
    "P6": (10,7), "P7": (3,3), "P8": (9,4), "P9": (3,7)
}
init_centroids = {"C1": (3,3), "C2": (3,7), "C3": (9,4)}

# Assignments from the distance table
assignments = {
    "P1": "C1", "P2": "C1", "P3": "C2", "P4": "C3", "P5": "C2",
    "P6": "C3", "P7": "C1", "P8": "C3", "P9": "C2"
}

# New centroids computed
new_centroids = {"C1": (2.00, 2.67), "C2": (3.67, 8.00), "C3": (9.00, 5.33)}

colors = {"C1": "red", "C2": "green", "C3": "blue"}

plt.figure(figsize=(8,6))

# Plot points colored by assigned cluster + labels
for name, (x, y) in points.items():
    c = assignments[name]
    plt.scatter(x, y, c=colors[c], s=100)
    plt.text(x+0.1, y+0.1, name, fontsize=9)

# Plot initial centroids (X) and new centroids (P)
for c, (x, y) in init_centroids.items():
    plt.scatter(x, y, c=colors[c], marker="X", s=180, label=f"{c} (init)")

for c, (x, y) in new_centroids.items():
    plt.scatter(x, y, c=colors[c], marker="P", s=160, label=f"{c} (new)")

plt.title("K-Means First Iteration: Initial vs New Centroids")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
