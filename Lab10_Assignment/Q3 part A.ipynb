```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Original 9 points + new point P10
points = {
    "P1": (1,3), "P2": (2,2), "P3": (5,8), "P4": (8,5), "P5": (3,9),
    "P6": (10,7), "P7": (3,3), "P8": (9,4), "P9": (3,7), "P10": (6,2)
}
X = np.array(list(points.values()))

# Run KMeans with K=3
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap="viridis", s=100)
plt.scatter(centroids[:,0], centroids[:,1], c="red", marker="X", s=200, label="Centroids")

# Label points
for i, label in enumerate(points.keys()):
    plt.text(X[i,0]+0.1, X[i,1]+0.1, label, fontsize=9)

plt.title("KMeans Clustering with P10 Added (K=3)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# Identify cluster of P10
p10_cluster = labels[-1]
print("P10 belongs to Cluster:", p10_cluster)

# Show new centroids
print("\nNew Centroid Locations:\n", centroids)
```
