```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
y = np.array([[0],[1],[1],[0]], dtype=np.float32)

activations = ['sigmoid', 'tanh', 'relu']
loss_curves = {}

for act in activations:
    print(f"\nActivation Function: {act}")
    model = models.Sequential([
        layers.Dense(4, input_dim=2, activation=act),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(X, y, epochs=300, verbose=0)
    loss, acc = model.evaluate(X, y, verbose=0)
    loss_curves[act] = history.history['loss']
    print(f"Loss: {loss:.4f}, Accuracy: {acc:.4f}")

plt.figure(figsize=(8,5))
for act in activations:
    plt.plot(loss_curves[act], label=act)
plt.title("Loss comparison by Activation Function (XOR Gate)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
