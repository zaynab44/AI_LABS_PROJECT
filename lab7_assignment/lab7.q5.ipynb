```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

notes = np.random.randint(0, 20, 200)

window = 10
X, y = [], []
for i in range(len(notes) - window):
    X.append(notes[i:i+window])
    y.append(notes[i+window])

X = np.array(X).reshape(-1, window, 1)
y = np.array(y)

model = Sequential([
    SimpleRNN(50),
    Dense(20, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
model.fit(X, y, epochs=20)

seed = notes[:10].reshape(1, window, 1)
predicted_note = np.argmax(model.predict(seed))

print("Generated note:", predicted_note)
```
