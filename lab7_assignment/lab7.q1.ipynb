```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
import numpy as np

text = """the sun is shining today the sun is bright today the moon is glowing tonight"""

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

input_sequences = []
words = text.split()

for i in range(3, len(words)):
    seq = words[i-3:i+1]
    encoded = tokenizer.texts_to_sequences([' '.join(seq)])[0]
    input_sequences.append(encoded)

input_sequences = np.array(input_sequences)

X = input_sequences[:, :-1]
y = input_sequences[:, -1]

model = Sequential([
    Embedding(total_words, 10, input_length=3),
    SimpleRNN(50),
    Dense(total_words, activation="softmax")
])

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(X, y, epochs=200, verbose=0)

seed = "the sun is"
token_list = tokenizer.texts_to_sequences([seed])[0]
token_list = pad_sequences([token_list], maxlen=3)

pred = model.predict(token_list)
predicted_word = tokenizer.index_word[np.argmax(pred)]

print("Next word:", predicted_word)
```
