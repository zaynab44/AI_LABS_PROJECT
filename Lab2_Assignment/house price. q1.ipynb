```python
# Q1: Multiple Linear Regression - House Price
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# --- Create a small toy dataset (or load CSV with pd.read_csv)
data = {
    "Size": [1500, 1700, 1200, 2000, 2300, 900, 1800, 1100],
    "Bedrooms": [3, 3, 2, 4, 4, 1, 3, 2],
    "Age": [10, 5, 20, 8, 2, 30, 12, 25],
    "Price": [300000, 350000, 210000, 420000, 480000, 160000, 380000, 200000]
}
df = pd.DataFrame(data)
X = df[["Size", "Bedrooms", "Age"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients (Size, Bedrooms, Age):", model.coef_)

# Interpret:
# coef[0] = change in price per 1 sqft,
# coef[1] = change in price per extra bedroom,
# coef[2] = change in price per extra year of age (likely negative)

# Predict for Size=2000, Bedrooms=3, Age=10
pred = model.predict(np.array([[2000, 3, 10]]))
print("Predicted price for (2000 sqft, 3 beds, age 10):", pred[0])
```
